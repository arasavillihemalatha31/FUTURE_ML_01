```python
# ✅ STEP 1: Install necessary libraries
!pip install transformers --quiet

from transformers import pipeline, set_seed

# ✅ STEP 2: Load the models
generator = pipeline("text-generation", model="distilgpt2")
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

set_seed(42)

# ✅ STEP 3: Show assistant menu
print("=== AI Assistant ===")
print("1. Answer a question")
print("2. Summarize text")
print("3. Generate creative content (story or poem)")
choice = input("Choose an option (1/2/3): ")

# ✅ STEP 4: Handle each option
if choice == "1":
    user_question = input("Enter your factual question: ")

    # Example context — in real use, fetch dynamically
    context = (
        "India is a country in South Asia. Its capital city is New Delhi, "
        "which serves as the seat of the government and the center of political activity."
    )

    result = qa_pipeline(question=user_question, context=context)
    response = result['answer']

elif choice == "2":
    user_input = input("Enter text to summarize: ")

    # Generate actual summary
    summary = summarizer(user_input, max_length=60, min_length=20, do_sample=False)
    response = summary[0]['summary_text']

elif choice == "3":
    user_input = input("Enter your story/poem prompt: ")

    response = generator(
        user_input,
        max_new_tokens=100,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.9,
        truncation=True
    )[0]["generated_text"]

else:
    print("Invalid choice.")
    exit()

# ✅ STEP 5: Show and save output
print("\n=== AI Response ===")
print(response)

feedback = input("\nWas this response helpful? (yes/no): ")

with open("feedback_log.txt", "a") as f:
    f.write(f"Choice: {choice}\nResponse: {response}\nFeedback: {feedback}\n\n")

print("\nThank you! Your feedback has been saved.")
```
