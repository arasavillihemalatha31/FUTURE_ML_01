```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
```

```python

np.random.seed(42)
dates = pd.date_range(start='2021-01-01', end='2022-12-31', freq='D')
sales = 200 + np.sin(np.linspace(0, 20, len(dates))) * 50 + np.random.normal(0, 15, len(dates))

df = pd.DataFrame({'ds': dates, 'y': sales})
df.head()
```

```python
# Weekly
df_weekly = df.resample('W-MON', on='ds').sum().reset_index()

# Monthly
df_monthly = df.resample('M', on='ds').sum().reset_index()

# Quarterly
df_quarterly = df.resample('Q', on='ds').sum().reset_index()

# Yearly
df_yearly = df.resample('Y', on='ds').sum().reset_index()
```

```python
def run_forecast(df, periods, freq_label, freq_alias):
    print(f"\n📈 Forecasting {freq_label.upper()} Sales")

    model = Prophet(
        daily_seasonality=(freq_alias == 'D'),
        weekly_seasonality=(freq_alias in ['D', 'W']),
        yearly_seasonality=True
    )

    model.fit(df)
    future = model.make_future_dataframe(periods=periods, freq=freq_alias)
    forecast = model.predict(future)

    # Plot forecast
    model.plot(forecast)
    plt.title(f"{freq_label.capitalize()} Sales Forecast")
    plt.xlabel("Date")
    plt.ylabel("Sales")
    plt.grid(True)
    plt.show()

    # Plot components
    model.plot_components(forecast)
    plt.show()
```

```python
# Daily forecast - 90 days
run_forecast(df, periods=90, freq_label='daily', freq_alias='D')

# Weekly forecast - 8 weeks
run_forecast(df_weekly, periods=8, freq_label='weekly', freq_alias='W')

# Monthly forecast - 6 months
run_forecast(df_monthly, periods=6, freq_label='monthly', freq_alias='M')

# Quarterly forecast - 4 quarters
run_forecast(df_quarterly, periods=4, freq_label='quarterly', freq_alias='Q')

# Yearly forecast - 2 years
run_forecast(df_yearly, periods=2, freq_label='yearly', freq_alias='Y')
```
